```python
from datasets import load_dataset
from datasets import ClassLabel, Dataset
from huggingface_hub import notebook_login
from transformers import AutoTokenizer, AutoModelForMaskedLM,AutoModelForSequenceClassification
import os
from typing import Dict, Union, List

import numpy as np
from sklearn.metrics import f1_score, recall_score, precision_score
from datasets import load_dataset
from datasets.formatting.formatting import LazyBatch
import evaluate
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
import torch
from transformers import TextClassificationPipeline
```

```python


dataset = load_dataset("mansoorhamidzadeh/snappfood")
```

```python


notebook_login()
```

```python
dataset['train']['label']
```

```python
id2label = {0: "HAPPY", 1: "SAD"}
label2id = {"HAPPY": 0, "SAD": 1}
```

```python

dataset = dataset.cast_column("label", ClassLabel(names=["HAPPY", "SAD"]))
```

```python
dataset['train']['label']
```

```python
# Load model directly


model_name = "PartAI/TookaBERT-Large"
tokenizer = AutoTokenizer.from_pretrained(model_name,force_download=True)

model = AutoModelForSequenceClassification.from_pretrained(
    model_name,
    num_labels=len(dataset['train'].features['label'].names),
    id2label=id2label,
    label2id=label2id
)
```

```python


def preprocess_function(examples: LazyBatch[str, List[List[str]]]) -> Dict:
    return tokenizer(
        examples["comment"],
        padding="max_length",
        truncation=True,
        max_length=512
    )


tokenized_dataset = dataset.map(
    preprocess_function,  remove_columns=['comment']
)
tokenized_dataset
```

```python

device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
```

```python
dataset['train'].features['label']
```

```python
def compute_metrics(eval_pred):
    prediction, labels = eval_pred
    prediction = np.argmax(prediction, axis=1)
    return {
        "accuracy": np.mean(prediction == labels),
        "f1-weighted": f1_score(labels, prediction, average='weighted'),
        "precision": precision_score(labels, prediction, average='weighted'),
        "recall": recall_score(labels, prediction, average='weighted')
    }
```

```python
model_name = "TookaBert_sentiment_snappfoood"

training_args = TrainingArguments(


    output_dir='./TookaBert_sentiment',
    logging_steps=1000,
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    gradient_accumulation_steps=32,
    evaluation_strategy="steps",
    save_total_limit=1,
    num_train_epochs=1,
    learning_rate=1e-5,
    push_to_hub=True,
)

```

```python
trainer = Trainer(
    model=model.to(device),
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["validation"],
    compute_metrics=compute_metrics,
    tokenizer=tokenizer
)
```

```python
trainer.train()
```

```python


classifier = TextClassificationPipeline(model=model, tokenizer=tokenizer)
```
