```python
from transformers import MT5ForConditionalGeneration,MT5Tokenizer
from datasets import load_dataset
from simpletransformers.t5 import T5Model, T5Args
from huggingface_hub import notebook_login, HfApi, HfFolder
```

```python
ds = load_dataset("persiannlp/parsinlu_translation_en_fa")
```

```python
ds.column_names
```

```python
train_data= ds['train']
test_set= ds['test']
validation_set= ds['validation']
```

```python
train_data[0]
```

```python
len(train_data)
```

```python
print(train_data[0]['source'])
print(train_data[0]['targets'][0])
```

```python
train_data_source=[]
train_data_targets=[]

for i in range(100000):
   train_data_source.append(train_data[i]['source'])
   train_data_targets.append(train_data[i]['targets'][0])
```

```python
print(len(train_data_source))
print(len(train_data_targets))
```

```python
print(train_data_source[3])
print(train_data_targets[3])
```

```python

prefix=[]
for i in range (100000):
   prefix.append('En to Fa translation')
```

```python
import pandas as pd
```

```python
ddata= [prefix,train_data_source,train_data_targets]
maindata= pd.DataFrame(zip(*ddata), columns=['prefix','input_text', 'target_text'])
```

```python
maindata.head()
```

```python
maindata.head()
```

```python
train_df= maindata[:95000]
eval_df = maindata [95000:100000]
```

```python
train_df
```

```python
model_args= T5Args()
```

```python
model_args.fp16= False
model_args.max_seq_length= 128
model_args.train_batch_size= 8
model_args.eval_batch_size = 8
model_args.num_train_epochs = 2
model_args.evaluate_during_training= True
model_args.evaluate_during_training_steps= 10000
model_args.use_multiprocessing= False
model_args.reprocess_input_data=True
model_args.overwrite_output_dir= True
model_args.save_steps=-1
model_args.no_cache= True
model_args.preprocess_inputs= False
model_args.num_return_sequences = 1
model_args.save_eval_checkpoints= False
model_args.max_length=128
model_args.num_beams=4
```

```python
model= T5Model ("mt5", "google/mt5-small",args=model_args, use_cuda=True)
```

```python
print(len(train_df))
print (len(eval_df))
```

```python
train_df.head()
```

```python
model.train_model(train_data=train_df,eval_data=eval_df)
```

```python
model.predict('this is a good day')
```
